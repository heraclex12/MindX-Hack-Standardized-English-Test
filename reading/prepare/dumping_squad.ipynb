```python
from pymongo import MongoClient
mongo = MongoClient('34.94.237.35', 27019)
db = mongo.mindX
collection = db.reading_bbc
```

```python
context = []
qa = []
with open('bbc_predict_input.txt', 'r', encoding='utf-8') as context_file:
    for line in context_file:
        context.append(line.strip().replace('squad: ', ''))
with open('bbc_predict_output.txt', 'r', encoding='utf-8') as qa_file:
    for line in qa_file:
        line = line.split('answer: ')
        if len(line) != 2:
            continue
        duo = []
        questions = line[0].replace('questions: ', '').strip()
        answers = line[1]
        
        for question, answer in zip(questions.split('|'), answers.split('|')):
            duo.append([question.strip(), answer.strip()])
        qa.append(duo)
    
data = []
for cont, q in zip(context, qa):
   data.append({'context': cont, 'qa': q})
```

```python
collection.insert(data)
```
